#### Use the appropriate visulalization to show the distribution of the column `Age`

```python
# your code here
plt.xlabel('Age')
plt.ylabel('Count')
plt.xticks(np.arange(0,100,5))
plt.hist(titanic['Age'], bins = len(titanic['Age'].unique()))
```

#### Use subplots to plot the distribution of the `Age` variable with bins equal to `10`, `20` and `50`.

```python
# your code here
fig, (ax1,ax2,ax3) = plt.subplots(3)
ax1,ax2,ax3.set_xlabel('Age')
ax1,ax2,ax3.set_ylabel('Count')

ax1.hist(titanic.Age, bins = 10)
ax2.hist(titanic.Age, bins = 20)
ax3.hist(titanic.Age, bins = 30)
```

#### Use the appropriate plot to visualize the column `Gender`. There are 2 ways of doing it. Do it both ways.

*Hint: Use matplotlib and seaborn*

```python
# your code here- 1st way
genders = titanic.groupby('Gender').agg('Gender').count()
plt.pie(genders,labels= ['MALE', 'FEMALE'], explode = [0,0.03], colors=['darkcyan','violet'])
```

```python
# your code here- 2nd way
sns.countplot(x='Gender', data=titanic, )
```

#### Use the appropriate plot to visualize the column `Pclass` 


```python
Pclasses = titanic.groupby('Pclass').agg('Pclass').count()
Pclasses
```

```python
# your code here
plt.title('Pclass count')
plt.pie(Pclasses,labels= ['1', '2', '3'], explode = [0,0,0.03], colors=['darkcyan','violet', 'darkorange'])
```
